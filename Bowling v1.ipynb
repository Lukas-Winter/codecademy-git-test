```python
# Klasse 'Game' beinhaltet die Spielsteuerung
class Game:
    def __init__(self, name):
        self.name = name
    
    def new_game(self):
        self.total_points = 0
        self.frame = 0
        self.frame_throws = {}
        
        for x in range(1,11):
            self.frame_throws[x] = Frame.throws(x)
            print("Würfe:", self.frame_throws)    
        
        self.total_points = Frame.total(self.frame_throws)
        
        print("Das Spiel ist vorbei.", self.name ,"hat insgesamt", self.total_points, "Punkte erzielt.")
```

```python
# Klasse 'Frame' beninhaltet die Berechnung der Würfe
class Frame:
    
    def throws(frame_num):
        frame_throws = []
        print("---------")
        print("Runde", frame_num)
        throw_one = Frame.verify_input()
        if throw_one == 10:
            frame_throws = [throw_one, 0]
        else:
            throw_two = Frame.verify_input()
            frame_throws = [throw_one, throw_two]
        if frame_num == 10 and (throw_one + throw_two) >= 10:
            throw_three = Frame.verify_input()
            frame_throws = [throw_one, throw_two, throw_three]
        return frame_throws
    
    
    def verify_input():
        verified = False
        while verified != True:
            try:
                txt = int(input("Wie viele Kegel hast du umgeworfen (0 - 10)?"))
                if 0 <= txt <= 10:
                    verified = True
                    return txt 
                else:
                    print("Gib bitte eine Zahl zwischen 0 und 10 an.")
                    txt = int(input("Wie viele Kegel hast du umgeworfen (0 - 10)?"))
            except ValueError:
                print("Gib bitte eine Zahl zwischen 0 und 10 an.")
    

    def total(frame_throws):
        points_frame = {}
        total = 0
        for x in frame_throws:
            if len(frame_throws[x]) > 2:
                points_frame[x] = frame_throws[x][0] + frame_throws[x][1] + frame_throws[x][2]
                total += points_frame[x]
            else:
                if frame_throws[x][0] == 10:
                    points_frame[x] = frame_throws[x][0] + frame_throws[x+1][0] + frame_throws[x+1][1]
                    total += points_frame[x]
                elif (frame_throws[x][0] + frame_throws[x][1]) == 10:
                    points_frame[x] = frame_throws[x][0] + frame_throws[x][1] + frame_throws[x+1][0]
                    total += points_frame[x]
                else:
                    points_frame[x] = frame_throws[x][0] + frame_throws[x][1]
                    total += points_frame[x]
        print("---------")
        print("Punkte pro Runde: ", points_frame)
        return total    
```
